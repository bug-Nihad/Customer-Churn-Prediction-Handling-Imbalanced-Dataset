```python
df = pd.read_csv('/kaggle/input/telecom-dataset/telco.csv')
df.head()
```

```python
df.drop('customerID', inplace=True, axis=1)
df.columns
```

```python
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
df.dtypes
```

```python
df = df.dropna(subset=['TotalCharges'])
df['TotalCharges'].isna().sum()
```

```python
def print_uniq_values(df):
    for col in df.columns:
        print(col, ": ", df[col].unique())
```

```python
df = df.replace({'No internet service': 'No',
                 'No phone service': 'No',
                'Yes': 1,
                 'No': 0,})
for col in df.columns:
    print(col, ": ", df[col].unique())
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

df[['TotalCharges', 'MonthlyCharges']] = scaler.fit_transform(df[['TotalCharges', 'MonthlyCharges']])
```

```python
df[['TotalCharges', 'MonthlyCharges']]
```

```python
df = pd.get_dummies(df)
```

```python
from sklearn.model_selection import train_test_split
X = df[[i for i in df.columns if i != 'Churn']]
Y = df['Churn']
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
X_train.shape, X_test.shape
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.metrics import classification_report


def ANN(X_train, y_train, X_test, y_test, epochs=10):
    model = Sequential()
    model.add(Dense(64, input_dim=X_train.shape[1], activation='relu'))
    model.add(Dense(32, activation='relu'))
    model.add(Dense(1, activation='sigmoid')) 

    model.compile(optimizer='adam',
                  loss='binary_crossentropy',
                  metrics=['accuracy'])
    model.fit(X_train, y_train,
                    epochs=epochs,  # Number of epochs
                    batch_size=32,  # Batch size
                    validation_split=0.2,  # Split part of the training data for validation
                    verbose=1)  # Show progress
    
    
    y_pred = model.predict(X_test)
    y_pred = y_pred.round().astype(int)
    report = classification_report(y_test, y_pred)
    print("Classification Report:")
    print(report)
    
    
```

```python
ANN(X_train, y_train, X_test, y_test, epochs=30)
```

# Undersampling

```python
df_undersampled = pd.concat([df[df['Churn']==0].sample(1869), df[df['Churn']==1]], axis=0)
df_undersampled['Churn'].value_counts()
```

```python
X = df_undersampled[[i for i in df_undersampled.columns if i != 'Churn']]
Y = df_undersampled['Churn']
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```
